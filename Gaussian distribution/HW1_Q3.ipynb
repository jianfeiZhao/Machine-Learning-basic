```python
from scipy.spatial import distance as dist
import numpy as np
import matplotlib.pyplot as plt

# d = 3
# generate 100 random samples from the standard multi-variate Gaussian distribution
mean1 = [0,0,0]
cov1 = [[1,0,0],[0,1,0],[0,0,1]]
data = np.random.multivariate_normal(mean1,cov1,10000)

dis = []
for i in range(100):
    for j in range(100):
        dis.append(dist.euclidean(data[i], data[j]))
_ = plt.hist(dis,bins='auto')
plt.title('Histogram of Euclidean norms')
plt.show()

print("Mean = {0:.2f}, Std = {1:.2f}".format(np.mean(dis), np.std(dis)))
```

```python
# d = 50
D = [50, 100, 200, 500, 1000]
for i in range(5):
 mean = np.zeros(D[i])
 cov = np.eye(D[i])
 data = np.random.multivariate_normal(mean,cov,10000)

 dis = []
 for a in range(100):
    for b in range(100):
        dis.append(dist.euclidean(data[a], data[b]))
 _ = plt.hist(dis,bins='auto')
 plt.title('Histogram of Euclidean norms')
 print("d = {0:}: Mean = {1:.2f}, Std = {2:.2f}".format(D[i], np.mean(dis), np.std(dis)))
```

```python
mean_a = [9.94, 13.85, 19.79, 31.13, 44.09]
std_a = [1.43, 1.73, 2.21, 3.28, 4.54]
_ = plt.scatter(D,mean_a,color='blue')
_ = plt.scatter(D,std_a,color='red')
plt.xscale('log')
plt.title('Blue = Means , Red = Standard deviations')
plt.show()
```
